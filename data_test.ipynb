```python
from pathlib import Path
from IPython import display
from natsort import natsorted
from PIL import Image
import matplotlib.pyplot as plt
import torchvision.transforms as T
```

```python
from src.dataloader import train_hr_transform, train_lr_transform
```

```python
hr_transform = train_hr_transform(512)
lr_transform = train_lr_transform(512, 4)

img_path = list(Path('data/train').glob('*.png'))
img_path = natsorted(img_path)
```

```python
import random


img = Image.open(img_path[random.randint(0, len(img_path))])
hr_img = hr_transform(img)
lr_img = lr_transform(hr_img)
hr_img = T.ToPILImage()(hr_img)
lr_img = T.ToPILImage()(lr_img)
```
